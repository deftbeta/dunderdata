# Missing Values
* Must handle missing values before doing any learning
* Simplest thing is to drop them from the dataset
* Numeric columns - replace with mean, median, or most frequent. Can also do machine learning with them.
* Categorical columns - replace with string 'missing' or most frequent.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

pd.options.display.max_columns = 100
%matplotlib inline

housing = pd.read_csv('data/kaggle_housing.csv')

# remove the sales price because we intend to predict it and to have it as predictor will
# result in a model that is always correct because the price is in the dataset

price = housing.pop('SalePrice')
housing.head()
```

```python
price.head()
```

```python
housing_numeric = housing.select_dtypes('number')
housing_numeric.head()
```

## A couple columns with missing values

```python
housing_numeric.isna().sum()

# this command looks for missing values and gives you a count of these for each col
```

```python
X = housing_numeric.values
y = price.values
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# attempt to fit
lr.fit(X, y)
```

### Must fill in missing values - Use `Imputer`
You can fill in the missing values with pandas, but scikit-learn has an Imputer estimator to automate the process for you.

```python
from sklearn.preprocessing import Imputer
```

```python
imp = Imputer(strategy='mean')
```

```python
X_filled = imp.fit_transform(X)
```

```python
np.isnan(X_filled).sum()
```

```python
lr.fit(X_filled, y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lr, X_filled, y, cv=5).mean()
```
