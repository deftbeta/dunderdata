### Objectives
This notebook will dive into the Seaborn visualization library in Python. This notebook assumes no previous visualization knowledge.

The [Python Seaborn library](http://seaborn.pydata.org/index.html) is a high-level, easy-to-use interface for creating powerful and beautiful visualizations. Seaborn does not actually perform any of the visualizations but instead, calls the primary Python visualization library, matplotlib, to do all the heavy lifting.

## Seaborn integration with Pandas
All Seaborn plotting functions contain a **`data`** parameter that accepts a pandas DataFrame. This allows you to use the string names of the column names for the other arguments in the plotting functions.

## Axes vs Grid Seaborn plots
All of seaborn plotting functions return either an 'Axes' or a 'Grid'. 

**Axes** are simpler plots that contain a single panel (a single x and y axis). They are technically matplotlib axes objects. Most seaborn plotting functions return axes.

**Grids** create multiple plots by adding dimensions to a plot. For instance, let's say we are creating bar charts that have the average salary of an employee by gender in a single plot. Seaborn Grids allow you to add another dimension, like race, to create these bar charts in separate plots for each race.

### Categorizing the Seaborn API
The main sections of [Seaborn's API][1] are
* Relational
* Categorical
* Distribution
* Regression

Each section contains a single Grid plot that is able to create all of the Axes plots. For instance, take a look at the categorical section of the API. The **`catplot`** function is the Grid function that simply can create all of the other Axes categorical plots in a grid. The Axes plots can only produce a single grid.

## Getting started with Axes plots for Univariate visualization
Let's begin by making plots with a single dimension of data.

[1]: http://seaborn.pydata.org/api.html

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
emp = pd.read_csv('data/employee.csv')
emp.head()
```

### Univariate numerical plots
A few of the plots work with just a single dimension of data. Pass the DataFrame to the **`data`** parameter. Let's create a boxplot. Most Seaborn methods have both an **`x`** and **`y`** parameter. For some plots you only need to specify one of them.

```python
sns.boxplot(x='salary', data=emp)
```

Make the boxplot vertical by passing using **`y`**.

```python
sns.boxplot(y='salary', data=emp)
```

### String Data
String data cannot be visualized with the above plots. We can assign the returned matplotlib Axes to a variable and then modify our plot. This is doing the same thing as `emp['RACE'].value_counts()` in Pandas.

```python
ax = sns.countplot('race', data=emp)
```

```python
ax = sns.countplot(y='race', data=emp)
```

### KDE and Histogram
The **`distplot`** function plots both the KDE and the histogram together. There is no **`data`** parameter here, so you must pass in the Series directly.

```python
sns.distplot(emp['experience'])
```

# Multivariate plotting
The above plots all contained a single plot. We will add another dimension to our data.

### Mixing a categorical feature
All the plots below use one categorical column and one numeric column. Seaborn will aggregate the data for us. Let's find the average salary by gender using many different plots.

```python
sns.boxplot(x='salary', y='gender', data=emp)
```

By default, the bar plot takes the average

```python
sns.barplot(x='gender', y='salary', data=emp)
```

## Multivariate numeric
We now turn to plotting two numeric variables with a scatter plot

The **`regplot`** function draws the regression line through the points.

```python
sns.regplot(x='experience', y='salary', data=emp)
```

# Add another dimension with `hue`
The `hue` parameter is found in most seaborn methods and allows you slice the data by one more dimension. The **`hue`** parameter does not create a new Axes, rather slices the data in the current Axes.

```python
sns.boxplot(x='salary', y='gender', hue='race', data=emp)
```

```python
sns.barplot(x='gender', y='salary', hue='race', data=emp)
```

### Add more Axes with the Grid plots
There are only a few Grid plots and for them, you will need to specify the name of the underlying Seaborn Axes plot you would like. **`catplot`** is one of the Grid plots.

## Use `col` or `row` to create the Grid
If you plot more than one Axes with Seaborn you have created a grid. Use the **`col`** or **`rol`** parameter to specify the column you would like to slice the data with.

The following plot has 

```python
sns.catplot(x='race', y='salary', hue='gender', data=emp, kind='bar', col='dept', col_wrap=3)
```
